```python
import requests
import subprocess
print('Start: 大明風華第35集.ts')    #    要更動     500
url='https://litvfreepc.akamaized.net/vod/content01/litv-repub1-vod42820-000035M001/litv-hls-cl-pc/litv-repub1-vod42820-000035M001-audio_eng=128360-video_eng=1439000-' #要更動
folder = r'c:\Users\User\Desktop\大明風華_測試'
f=open(folder+'\\'+'大明風華第35集__.ts','ab')   #要更動 
lastLen=0
Duration= (46-1) * 60 + 0
N=int(Duration/9 )
print('N=',N) 
for j in range(1, 2200):  #要更動 
    url_list=url+str(j)+'.ts'
    r=requests.get(url_list,stream=True)   #發出請求
    if r.status_code == 200:  # 如果回傳不是找不到網頁
        r=r.content 
        L=len(r)
        if L!=lastLen: 
            f.write(r)
            lastLen=len(r)
            print('j=',j,', 長=',lastLen,end="") 
            if L!=lastLen: 
                print('同前?',False,end="")
            print()
            
        else: 
            print("(跟前一個一樣長，跳出)")
            break 
    else: 
        break 

    if j%10==0:
        print("")
    if j%100==0:
        print("")
    
f.close() 
print('\n\n結束。')

```
